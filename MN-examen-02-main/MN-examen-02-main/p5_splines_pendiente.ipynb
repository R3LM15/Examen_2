# Splines cúbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Tarea
* Grafique los splines cúbicos tal que cumplan con una pendiente de $m=-3$ en los puntos dados.

```python
import matplotlib.pyplot as plt
import numpy as np

def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

xs = [-1, 0, 1]
ys = [1, 5, 3]
# Coeficientes ajustados para que la pendiente en x=0 sea -3
s = [
    {"a": 1, "b": 9, "c": 4, "d": -4},     # S0 para [-1, 0]
    {"a": 5, "b": -3, "c": -8, "d": 6},    # S1 para [0, 1]
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)

x_tan = np.linspace(-0.5, 0.5, 10)
y_tan = 5 + (-3) * (x_tan - 0) 
plt.plot(x_tan, y_tan, 'b--', label="Pendiente deseada")

plt.text(0.15, 4.7, r"$m_{deseada}$", fontsize=14, style='italic')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.legend()
plt.show()
```
